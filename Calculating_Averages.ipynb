```python
import pandas as pd

def count_values(frame, columns=[]):
    def calculate_count(previous, column):
        return previous + frame[column]
    return reduce(calculate_count, columns, 0)

def percents(hits, outs, walks):
    total = (hits + walks + outs)
    hitting_average = None
    walking_average = None
    if (total is not 0):
        hitting_average = float(hits) / total
    if (total is not 0):
        walking_average = float(walks) / total
    return hitting_average, total, walking_average

def percents_for_prefix(frame, prefix):
    frame = frame.fillna(0)
    hits = count_values(frame, [prefix + '20', prefix + '21', prefix + '22', prefix + '23'])
    walks = count_values(frame, [prefix + '14', prefix + '15', prefix + '16', prefix + '17'])
    outs = count_values(frame, [prefix + '2', prefix + '3', prefix + '18', prefix + '19'])
    hitting_average, total, walking_average = percents(hits, outs, walks)
    return hitting_average, total, walking_average

def calculate_hitting_percent(frame):
    hitting_average, total, walking_average = percents_for_prefix(frame, 'previous_batter_')

    return pd.Series([total, hitting_average, walking_average], index=['total_batter_appearances', 'hitting_average', 'walking_average'])


def calculate_pitching_percent(frame):
    hitting_average, total, walking_average = percents_for_prefix(frame, 'previous_pitcher_')

    return pd.Series([total, hitting_average, walking_average], index=['total_pitcher_appearances', 'pitcher_hitting_average', 'pitcher_walking_average'])


def calculate_pitching_against_hand_percent(frame):
    hitting_average, total, walking_average = percents_for_prefix(frame, 'pitcher_against_hand_')

    return pd.Series([total, hitting_average, walking_average], index=['hand_pitcher_appearances', 'hand_pitcher_hitting_average', 'hand_pitcher_walking_average'])


def calculate_batter_against_hand_percent(frame):
    hitting_average, total, walking_average = percents_for_prefix(frame, 'batter_against_hand_')

    return pd.Series([total, hitting_average, walking_average], index=['hand_batter_appearances', 'hand_hitting_average', 'hand_walking_average'])


def calculate_bullpen_percent(frame):
    hitting_average, total, walking_average = percents_for_prefix(frame, 'bullpen_')

    return pd.Series([total, hitting_average, walking_average], index=['bullpen_appearances', 'bullpen_average', 'bullpen_walking_average'])
```

```python
choices = choices.join(choices.apply(calculate_hitting_percent, axis=1)
.join(choices.apply(calculate_pitching_percent, axis=1))
.join(choices.apply(calculate_pitching_against_hand_percent, axis=1))
.join(choices.apply(calculate_batter_against_hand_percent, axis=1))
.join(choices.apply(calculate_bullpen_percent, axis=1)))
```

```python

no_hit = pd.Series([0], index=['got_hit'])
hit = pd.Series([1], index=['got_hit'])
def calculate_got_hit(frame):
    if int(frame['best_hit']) > 0:
        return hit
    else:
        return no_hit

choices = choices.join(choices.apply(calculate_got_hit, axis=1))
```
